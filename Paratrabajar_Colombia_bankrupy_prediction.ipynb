```python

# Imputar los valores faltantes con la media de cada columna
for col in ["Costo de ventas", "Ganancias acumuladas", "Pasivos corrientes totales", "Activos corrientes totales"]:
    train_df[col] = train_df[col].fillna(train_df[col].mean())
    test_df[col] = test_df[col].fillna(test_df[col].mean())

# Codificar la columna `Sector` usando one-hot encoding
X_train = pd.get_dummies(train_df, columns=["Sector"])
X_test = pd.get_dummies(test_df, columns=["Sector"])

# Extraer la variable objetivo
y_train = train_df["event"]

# Eliminar columnas innecesarias
X_train.drop(columns=["Costo de ventas", "index", "event"], inplace=True)
X_test.drop(columns=["Costo de ventas", "index"], inplace=True)
```

```python

# Definir los pipelines
pipelines = {
    "Escalado Min-Max": Pipeline([
        ("escalador", MinMaxScaler()),
        ("clasificador", DecisionTreeClassifier(random_state=42))
    ]),
    "Escalado Estándar": Pipeline([
        ("escalador", StandardScaler()),
        ("clasificador", DecisionTreeClassifier(random_state=42))
    ]),
    "Normalización L2": Pipeline([
        ("escalador", Normalizer(norm="l2")),
        ("clasificador", DecisionTreeClassifier(random_state=42))
    ])
}

# Crear un diccionario para almacenar las puntuaciones de validación cruzada
puntuaciones = {}

# Definir múltiples métricas de evaluación
metricas_evaluacion = {
    'f1': 'f1',
    'exactitud': 'accuracy',
    'sensibilidad': 'recall',
    'AUC': 'roc_auc'
}

# Iterar sobre los pipelines y calcular las puntuaciones de validación cruzada
for nombre, pipeline in pipelines.items():
    resultados_cv = cross_validate(pipeline, X_train, y_train, cv=5, scoring=metricas_evaluacion, return_train_score=False)

    # Almacenar las puntuaciones medias para cada métrica
    puntuaciones[nombre] = {metrica: np.mean(resultados_cv[f'test_{metrica}']) for metrica in metricas_evaluacion}
    print(f"{nombre} - Puntuaciones de validación cruzada:")
    for metrica, valor in puntuaciones[nombre].items():
        print(f"   {metrica.capitalize()}: {valor:.10f}")


```

```python

# Convertir el diccionario de puntuaciones en un DataFrame
puntuaciones_df = pd.DataFrame(puntuaciones).T

# Calcular el promedio de las métricas para cada pipeline y agregarlo al DataFrame
puntuaciones_df['puntuacion_promedio'] = puntuaciones_df.mean(axis=1)

# Seleccionar el pipeline con la puntuación promedio más alta
nombre_mejor_pipeline = puntuaciones_df['puntuacion_promedio'].idxmax()
mejor_pipeline = pipelines[nombre_mejor_pipeline]

# Entrenar el mejor pipeline en todo el conjunto de entrenamiento
mejor_pipeline.fit(X_train, y_train)

# Generar predicciones en el conjunto de prueba
y_pred = mejor_pipeline.predict(X_test)
```
